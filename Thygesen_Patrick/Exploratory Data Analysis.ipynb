# Salary Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('CSC485/CSC485-Sp24/Thygesen_P/salary_data_cleaned.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
average_salary_by_ownership = df.groupby('Type of ownership')['avg_salary'].mean().reset_index()
print(average_salary_by_ownership)
```

```python
plt.figure(figsize=(7,))
plt.bar(average_salary_by_ownership['Type of ownership'], average_salary_by_ownership['avg_salary'], color = 'pink')
plt.xlabel('Type of Ownership')
plt.ylabel('Average Salary')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
average_salary_by_sector = df.groupby('Sector')['avg_salary'].mean().reset_index()
print(average_salary_by_sector)
```

```python
plt.figure(figsize=(7,4))
plt.bar(average_salary_by_sector['Sector'], average_salary_by_sector['avg_salary'], color = 'pink')
plt.xlabel('Sector')
plt.ylabel('Average Salary')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Some questions for this dataset could be:
Which state has the highest salaries in a certain industry (i.e. Data Science)?
Which sector has the highest average rating?
Which type of company has a higher average rating?
Which industry's have the oldest workers?

# Shopping Trends Dataset

```python
df2 = pd.read_csv('CSC485/CSC485-Sp24/Thygesen_P/shopping_trends_updated.csv')
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2.describe()
```

```python
total_sales_by_items = df2.groupby('Item Purchased')['Purchase Amount (USD)'].sum().reset_index()
print(total_sales_by_items)
```

```python
plt.figure(figsize=(7,4))
plt.bar(total_sales_by_items['Item Purchased'], total_sales_by_items['Purchase Amount (USD)'], color = 'pink')
plt.xlabel('Item')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
total_sales_by_season = df2.groupby('Season')['Purchase Amount (USD)'].sum().reset_index()
print(total_sales_by_season)
```

```python
plt.figure(figsize=(7,4))
plt.bar(total_sales_by_season['Season'], total_sales_by_season['Purchase Amount (USD)'], color = 'blue')
plt.xlabel('Season')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Questions to answer from this data Set:
Which item of clothing sells the most in certain seasons?
Which state has the highest sales?
How does subscription status correlate with review ratings?

# Housing Data

```python
df3 = pd.read_csv('CSC485/CSC485-Sp24/Thygesen_P/NY-House-Dataset.csv')
df3.head()
```

```python
df3.isnull().sum()
```

```python
df3.describe()
```

```python
average_price_by_sublocality = df3.groupby('SUBLOCALITY')['PRICE'].mean().reset_index()
print(average_price_by_sublocality)
```

```python
plt.figure(figsize=(7,4))
plt.bar(average_price_by_sublocality['SUBLOCALITY'], average_price_by_sublocality['PRICE'], color = 'pink')
plt.xlabel('SUBLOCALITY')
plt.ylabel('PRICE')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
average_price_by_type = df3.groupby('TYPE')['PRICE'].mean().reset_index()
print(average_price_by_type)
```

```python
plt.figure(figsize=(7,4))
plt.bar(average_price_by_type['TYPE'], average_price_by_type['PRICE'], color = 'green')
plt.xlabel('TYPE')
plt.ylabel('PRICE')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Some possible questions to further explore with this dataset would be:
How much an additional bed and bath cost in a house - correlation?
How do the size of property's differ by each locality?
